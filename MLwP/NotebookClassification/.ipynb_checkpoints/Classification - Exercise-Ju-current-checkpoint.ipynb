```python
## 3 Exercise
The actual exercise is divided into 5 tasks:
* 3.1 **Getting Hands on the Data**
* 3.2 **Logistic Regression**
* 3.3 **Gradient Descent**
* 3.4 **Accuracy - How well did we do?**
* 3.5 **Multiclass Classification (One vs All)**

Your task is to fill in `...` under `### STUDENT TASK ###` in each step.

## 3.1 Getting Hands on the Data
After reading in the RGB images, we have to change them into a form that we can process more easily. Although the images are quite small (around $3000 \times 3000$ pixels), it is better to not work represent them with a long vector obtained from stacking the pixel colour values into a vector due to limited computational resources. Instead, we represent an  image by three features $\mathbf{x}=(x_{r},x_{g},x_{b})^{T} \in \mathbb{R}^{3}$ obtained by averaging the redness, greenness and blueness over all pixels. Let us stack the feature vectors $\mathbf{x}^{(i)} \in \mathbb{R}^{3}$, for $i=1,\ldots,N$, obtained for all the training images into the "feature matrix" $\mathbf{X}$

\begin{equation*}
    \mathbf{X} = \big(\mathbf{x}^{(1)},\dots,\mathbf{x}^{(55)}\big)^T=\begin{bmatrix}
    x^{(1)}_{r}  & x^{(1)}_{g}  & x^{(1)}_{b} \\
    \vdots & \ddots & \vdots\\
    x^{(55)}_{r} & x^{(55)}_{g} & x^{(55)}_{b}
    \end{bmatrix}
    \label{xm}
    \tag{1}
\end{equation*}

Here, $x^{(1)}_{r}$ denotes the average red component of the first image in the trainging dataset. 

Each image in the training set has a label $y^{(i)}$ where $y^{(i)} = 1$ if the i-th image shows grass while $y^{(i)} = 0$ otherwise (when it shows soil or tiles). It will be convenient to stack the labels into the "label vector"

\begin{equation*}
    \mathbf{y}=\begin{bmatrix}
    y^{(1)}\\
    y^{(2)}\\
    \vdots\\
    y^{(N)}
    \end{bmatrix}\in \mathbb{R}^{N}
    \label{vy}
    \tag{2}
\end{equation*}

In this task you are required to 
- Implement a Python function `get_feature_matrix()` which returns the feature matrix (\ref{xm}).
- Implement a Python function `get_labels()`, returning the label vector (\ref{vy}).
- Implement a Python function `Visualize_data()` to draw 3 **scatter plots**
    - for each color combination (*Greenness vs Redness*,*Greenness vs. Blueness* and *Redness vs Blueness*)
        - Mark a datapoint with cross if it's a grass picture or dot if not
    - Plot labels and axes are preconfigured, your job is to setup correct x and y -values to the each `plt.scatter()`-function.

???? Tips:
- You might want to save the feature matrix $\mathbf{X}$ as .txt, .csv or any other format that is convenient for you. This way you can just reload the matrix from the file without processing the images again.????
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
from numpy.linalg import inv
import math

def get_feature_vector(N = 55):

    x_vec = np.zeros((N,3))
    print (x_vec.shape)

    x = []
    ### STUDENT TASK ###
    ## Loop through each picture and each pixel and sum RGB values into the feature vector matrix.
    ## At last, remember to divide each R, G and B sum with total pixel count.
    ## Hints: Most of the commands required for this task are in the 2 Dataset -section.
    for i in range(N):
        sum_r = 0; sum_g = 0; sum_b = 0
        im = Image.open("images/image_{}.jpg".format(i+1))
        width, height = im.size # Size of Image
        total_pixels = width*height
        rgb_im = im.convert('RGB')


        for row in range(width):
            for col in range(height):
                r,g,b = rgb_im.getpixel((row,col))
                sum_r += r
                sum_g += g
                sum_b += b
                
        x_vec[i,0] = sum_r/total_pixels
        x_vec[i,1] = sum_g/total_pixels
        x_vec[i,2] = sum_b/total_pixels

    return x_vec;

def get_labels(N=55):
    y = np.zeros((N,1));
    ### STUDENT TASK ###
    ## Generate label vector where 1 is Grass image and 0 is Non-Grass
    ## Hints: See 2 Dataset -section where the picture order is defined
    for i in range(N):
        if (i < 20):
            y[i,:] = 1 # Grass
        else:
            y[i,:] = 0 #Non-Grass
        #else:
        #y[i,:] = 3
    return y

""" VISUALIZE THE DTATA """
def Visualize_data(X,y):
    # PLOT GREENNESS AGAINST REDNESS
    ### STUDENT TASK ###
    indx_1 = np.where(y == 1) # for grass.
    indx_2 = np.where(y == 0) # for non-grass.
    
    # Set figure size (width, height)
    fig, axes = plt.subplots(1, 3,figsize=(15, 5))
    ### STUDENT TASK ###
    # l1 = plt.scatter(..., ..., c='g', marker ='x')
    # l2 = plt.scatter(..., ..., c='r', marker ='o')
    axes[0].scatter(X[indx_1,1], X[indx_1,0], c='g', marker ='x', label='Grass')
    axes[0].scatter(X[indx_2,1], X[indx_2,0], c='r', marker ='o', label='Soil+Tiles')
    
    axes[0].set_xlabel('Greenness of Images')
    axes[0].set_ylabel('Redness of Images')
    axes[0].legend()
    axes[0].set_title(r'$\bf{Figure\ 1.}$Green vs Red')

    # PLOT GREENNESS AGAINST BLUENESS
    ### STUDENT TASK ###
    # l1 = plt.scatter(..., ..., c='g', marker ='x')
    # l2 = plt.scatter(..., ..., c='b', marker ='o')
    axes[1].scatter(X[indx_1,1], X[indx_1,2], c='g', marker ='x', label='Grass')
    axes[1].scatter(X[indx_2,1], X[indx_2,2], c='b', marker ='o', label='Soil+Tiles')
    
    axes[1].set_xlabel('Greenness of Images')
    axes[1].set_ylabel('Blueness of Images')
    axes[1].legend()
    axes[1].set_title(r'$\bf{Figure\ 2.}$Green vs Blue')

    # PLOT REDNESS AGAINST BLUENESS
    ### STUDENT TASK ###
    # l1 = plt.scatter(..., ..., c='g', marker ='x')
    # l2 = plt.scatter(..., ..., c='b', marker ='o')
    axes[2].scatter(X[indx_1,0], X[indx_1,2], c='r', marker ='x', label='Grass')
    axes[2].scatter(X[indx_2,0], X[indx_2,2], c='b', marker ='o', label='Soil+Tiles')
    
    axes[2].set_xlabel('Redness of Images')
    axes[2].set_ylabel('Blueness of Images')
    axes[2].legend()
    axes[2].set_title(r'$\bf{Figure\ 3.}$Red vs Blue')
    plt.tight_layout()
    return axes
```

```python
## 3.2 Logistic Regression
Our goal is to determine the label $y\in \{0,1\}$ (y=1 for "Grass" and y=0 for "No Grass2) of a snapshot based on the features $\mathbf{x}=(x_{r},x_{g},x_{b})^{T}$ given by average redness, greenness and blueness. In principle we can use a linear function $h^{(\mathbf{w})}(x)= \mathbf{w}^{T} \mathbf{x}$ to approximate the label $y$ based on the features $\mathbf{x}$. However, the predicted label $h^{(\mathbf{w})}(\mathbf{x}$ is typically not a value in the set $\{0,1\}$. Therefore, we need to quantize the predicted label according 

\begin{equation*} 
    \hat{y} = \begin{cases} 
        1 &\text{if}\ h^{(\mathbf{w})}(\mathbf{x}) \geq 1/2\\
        0 &\text{if}\ h^{(\mathbf{w})}(\mathbf{x}) < 1/2.
    \end{cases}
    \label{eq_classify}
    \tag{6}
\end{equation*}

Moreover, in contrast to linear regression, we use the logistic loss to measure the quality of a classifier $h^{(\mathbf{w})}$ obtained for a particular choice of the weight vector $\mathbf{w} \in \mathbb{R}^{3}$. The logistic loss is defined as:

\begin{equation*}
    \mathcal{L}\big((\mathbf{x},y),h^{(\mathbf{w})}(\cdot)\big) = -y^T\log\big(h^{(\mathbf{w})}(\mathbf{x})\big)-(1-y)^T\log\big(1-h^{(\mathbf{w})}(\mathbf{x})\big)
    \label{loss}
    \tag{3}
\end{equation*}

with
\begin{equation*}
    \sigma(\mathbf{w}^Tx) = \frac{1}{1+exp(-\mathbf{w}^T\mathbf{x})}=\frac{1}{1+e^{-\mathbf{w}^T\mathbf{x}}}.
    \label{sigmoid}
    \tag{4}
\end{equation*}


Since we have N=55 labeled images, with features $\mathbf{x}^{(i)}$ and the true label $y^{(i)}$, we can evaluate the loss of a classifier $h^{(\mathbf{w})}$ for all those images to obtain the empirical risk (or, average loss)
\begin{equation*}
\mathcal{E}(\mathbf{x}\mid\mathcal{X}) = \frac{1}{N} \sum_{i=1}^{N} \mathcal{L}((\mathbf{x}^{(i)},y^{(i)}),\ h^\mathbf{w}(\mathbf{x})  = \sigma(\mathbf{w}^Tx)
\end{equation*}
Note that for a given (fixed) dataset $\mathcal{X}$, the empirical risk $\mathcal{E}( \mathbf{w} \mid \mathcal{X})$ is a differentiable convex function $f(\mathbf{w}) =\mathcal{E}( \mathbf{w} \mid \mathcal{X})$ of the weight vector $\mathbf{w}$. Therefore, we can use **gradient descent (GD)** to find the optimal weight vector which minimizes the empirical risk. In particular, GD constructs a sequence of weight vectors $w^{(k)}$ by iterating (repeating) the GD update

\begin{equation*}
    \mathbf{w}^{(k+1)} = \mathbf{w}^{(k)} âˆ’ \alpha\nabla \mathcal{E}(\mathbf{w}^{(k)}\mid\mathcal{X})
    \label{gd}
    \tag{5}
\end{equation*}

Assume we run the GD updates for $k$ iterations which results in the weight vector $\mathbf{w}^{(k)}$ and corresponding classifier map $h^{(\mathbf{w}{(k)})}(\mathbf{x}) = \big(\mathbf{w}^{(k)} \big)^{T} \mathbf{x}$. Using the classifier map, we can compute the predicted label 



### Tasks
- implement a Python function `sigmoid_function()` (\ref{sigmoid}) which takes a dot weight vector and feature vector as inputs and returns the calculated value from sigmoid function.
- impleent a Python function 'gradient()' which reads in the feature matrix, label vector, a weight vector $\mathbf{w}$ and returns the gradient of the empirical risk at the weight $\mathbf{w}$
- implement a Python function `logisticRegression_func()` which takes as input the feature matrix $\mathbf{X}$, the label vector $\mathbf{y}$ and the number $K$ of GD updates to be run. The function should deliver the final GD iterate $\mathbf{w}^{(K)}$, a vector of length $N$ containing the predicted labels $\hat{y}^{(i)}$ (see Eq. (\ref{eq_classify}) and a vector of length $K$ whose $k$th entry is the empirical risk $\mathcal{E}(\mathbf{w}^{(k)}\mid\mathcal{X})$
- implement a Pyton function `classify()
```

```python
def sigmoid_func(z):
    ### STUDENT TASK ###
    # sigmoid = ...
    ### BEGIN SOLUTION
    sigmoid = 1/(1+np.exp(-z));
    ### END SOLUTION
    return sigmoid

def gradient(X,y,N,z):
    ### STUDENT TASK ###
    # grad = ...
    ### BEGIN SOLUTION
    grad =  1/N*np.dot(X.T,sigmoid_func(z)-y)
    ### END SOLUTION
    return grad

def logisticRegression_func(X,y,step_size, num_iter):
    N = X.shape[0]
    d = X.shape[1]
    # Initialize w as 1xd array.
    w = np.zeros((1,d))
    loss = float('inf')
    loss_list = []
    for i in range(num_iter):
        ### STUDENT TASK ###
        ### BEGIN SOLUTION
        z = np.dot(X,w.T)
        grad = gradient(X,y,N,z)
        w = w - step_size*grad.T # 1xd
        z = np.dot(X,w.T) # Nx1
        h = sigmoid_func(z) # Nx1
        loss_1 = - np.dot( y.T, np.log(h) ) # 1x1 y^T*log(h(z))
        loss_0 = - np.dot( (np.ones((N,1))-y).T, np.log(np.ones((N,1))-h) ) # 1x1 (1-y)^T*log(1-h(z))
        loss = loss_1 + loss_0
        loss_list.append(loss[0])
        ### END SOLUTION
    return loss_list, w

""" Predict Output """
def predict_output(X,w):
    ### STUDENT TASK ###
    ### BEGIN SOLUTION
    z = np.dot(X,w.T)
    predict_y = sigmoid_func(z)
    N = predict_y.shape[0]
    y = np.zeros((N,1));
    for i in range(N):
        if (predict_y[i,:]>=0.5):
            y[i,:] = 1;
        elif (predict_y[i,:]<0.5):
            y[i,:] = 0;
    ### END SOLUTION
    return y
```

```python
# Execute this cell. Do not modify.
step_size = 1e-5
num_iter = 3000
e_list, w_opt = logisticRegression_func(X,y,step_size,num_iter)
print (f'Weight Vector is: {w_opt}')
y_hat = predict_output(X,w_opt)

assert np.sum(e_list)/num_iter < 40
assert y_hat.shape == (55,1)
y_predict_test = predict_output(np.array([[-1,1],[0,1],[-1,1],[1,1]]),np.array([[1,0]]))
assert y_predict_test.shape == (4,1) 
np.testing.assert_array_equal(y_predict_test, np.array([[0],[1],[0],[1]]))

# Test prediction function
for i in np.linspace(-1,1,100,endpoint=False):
    if i < 0:
        np.testing.assert_array_equal(predict_output(np.array([[i,1]]),np.array([[1,0]])),np.array([[0]]), "Incorrect output. Check that you defined and interpreted sigmoid correctly.")
    else:
        np.testing.assert_array_equal(predict_output(np.array([[i,1]]),np.array([[1,0]])),np.array([[1]]), "Incorrect output. Check that you defined and interpreted sigmoid correctly.")

### BEGIN HIDDEN TESTS
from unittest.mock  import patch
# Test that each function is called
with patch(f"__main__.sigmoid_func",side_effect=sigmoid_func) as mock:
    logisticRegression_func(np.array([[0,1]]),np.array([0]),step_size,200)
    assert mock.called
with patch(f"__main__.gradient",side_effect=gradient) as mock:
    logisticRegression_func(np.array([[0,1]]),np.array([0]),step_size,200)
    assert mock.called

### END HIDDEN TESTS
```

```python
## 3.4 Accuracy - How well did we do?
Calculate the accuracy achieved the classifier $h^{(\mathbf{w})}$ obtained from 'Task 3.3 - Gradient Descent' by dividing the amount of correctly labeled pictures with total picture count, i.e.
\begin{equation*}
    \text{Accuracy} =\dfrac{1}{N} \sum_{i=1}^{N} \mathcal{I}(\hat{y}^{(i)} \neq y^{(i)})
    \label{acc}
    \tag{5}
\end{equation*}



### Tasks
- Implement a Python function `calculate_accuracy()` with takes two input argumetns the vector of true labels and the vector of predicted labels and returns the accuracy according to (Eq. \ref{acc}).
```

```python
def calculate_accuracy(y,y_hat):
    ### STUDENT TASK ###
    ### BEGIN SOLUTION
    N = y.shape[0];
    correct = 0;
    for i in range(N):
        if (y[i]==y_hat[i]):
            correct += 1
    accuracy = round(correct/N,4)*100
    ### END SOLUTION
    return accuracy

print (f'Accuracy of the result is: {calculate_accuracy(y,y_hat)}%')
```

## 3.5: Multiclass Classification (One vs All)
Instead of classifying the current on-board image as showing grass or not, we now consider how to classify the image according to the three categories "grass", "soil" or "tiles". 

So is all of our previous work on classifying images into "grass" vs. "no grass" for nothing? Nope! Adapting a binary classification method (using two different label values) to this multiclass task is straightforward using the **â€œone vs allâ€ technique**. The idea is quite simple: split the multiclass problem into three subproblems, each subproblem being to a binary classification problem as considered in task `3.3 Logistic Regression`. In particular, we can solve the problem of classifiying images into classes "grass", "soil" or "tiles" into the three subproblems:
* subproblem 1: classifying images according to "grass"(y=1) vs. "no grass"(y=0); 
* subproblem 2: classifying images according to "soil"(y=1) vs. "no soil"(y=0)
* subproblem 3: classiying images according to "tiles"(y=1) vs. "no tiles"(y=0)

The first subproblem has already be solved in Sec. 3.3, so we only need to solve the latter two subproblems. 
Let us illustrate this with an example: Assume we obtain the following predictor values for the three weight vectors applied to an image with features x: 
- $h^{(w_{grass})}(x) = 2$ ("grass vs. no grass")
- $h^{(w_{soil})}(x) = 1$ ("soil vs. no soil") 
- $h^{(w_{tiles})}(x) = 5$ ("tiles vs. no tiles")

Then we classify this image as tiles.
 
Essentially we use our `3.2: Logistic Regression` solution 3 times by reassigning label vector ($y$) according to class we are going to predict (ie. altering which of pictures [Grass, Soil or Tiles] are assigned as 1).

### Tasks
- Implement a Python function "multiclassifier()" which implements logistic regression for the three subproblems and then outputs the final predicted labels
- the function should return the predicted labels in a vector whose $i$th entry is 0 if the $i$th image is classified as grass, is 1 if the $i$th image is classified as "Soil" and is equal to 2 if the $i$th image is classified as "Tiles". 
- The function should also return the accuracy as in `3.4: Accuracy - How well did we do?`

```python
def get_labels_k(N=55,k=0):
    y = np.zeros((N,1));
    ### STUDENT TASK ###
    ## Generate label vector where 1 is Grass image and 0 is Non-Grass
    ## Hints: See 2 Dataset -section where the picture order is defined
    ### BEGIN SOLUTION
    for i in range(N):
        if (k*20 <= i < (k+1)*20):
            y[i,:] = 1 # Grass / Tiles / Soil
        else:
            y[i,:] = 0 #Non-Grass / Non-Tiles / Non-Soil
    ### END SOLUTION
    return y

def multiclass():
    y_predict = []
    step_size = 1e-5
    num_iter = 3000
    for i in range(0,3):
        ### STUDENT TASK ###
        ### BEGIN SOLUTION
        _, w_opt = logisticRegression_func(
            #X=...,
            #y=...,
            X=X,
            y=get_labels_k(k=i),
            step_size=1e-5,
            num_iter=3000)
        y_predict_i = sigmoid_func(np.dot(X,w_opt.T))
        N = y_predict_i.shape[0]
        if not len(y_predict):
            y_predict = np.zeros((N,3))
            y_predict[:,i] = y_predict_i.reshape(55)
        else:
            y_predict[:,i] = y_predict_i.reshape(55)
    y = np.array([0]*20+[1]*20+[2]*15)
    y_hat = y_predict.argmax(axis=1)
    accuracy = calculate_accuracy(y, y_hat)
    ### END SOLUTION
    return y_hat, accuracy
```

## DEMO: Confusion Matrix

In task `3.5 Multiclass Classification (One vs All)` we have implemented a classifier which associates an image with feature vector $x$ to one of the three categories: "Grass", "Soil" and "Tiles".

We can evaluate the quality of the classifier by determining the confusion matrix. The 
entry in the ith row and jth column of the confusion matrix is 
the fraction of images which have true label $y=i$ but are classified to label $\hat{y}=j$. 


http://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

**NOTE:**

http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py

**Takes too much time to create yourself. For this DEMO to work properly, you need to complete task `3.5`**

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
class_names = ['Grass', 'Soil','Tiles']
cnf_matrix = confusion_matrix(y, y_hat)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure(figsize=(12, 4))
plt.subplot(121) 
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.subplot(122) 
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```
